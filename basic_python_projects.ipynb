# 2 -  Projet myst√®re

Le but de ce projet est de permettre √† un joueur d'essayer de deviner un nombre myst√®re g√©n√©r√© al√©atoirement par l'ordinateur, en 5 essais ou moins.

**D√©roul√© du script**

Au d√©but du script, vous devez g√©n√©rer un nombre al√©atoire compris entre 0 et 100 (vous pouvez agrandir ou r√©duire cet intervalle pour simplifier ou complexifier le jeu).

Le joueur dispose alors de 5 essais (l√† encore, libre √† vous de changer cette valeur) pour trouver le nombre myst√®re.

√Ä chaque essai, vous devez indiquer au joueur si le nombre qu'il a entr√© est plus petit ou plus grand que le nombre myst√®re.

Si le nombre entr√© par l'utilisateur est √©gal au nombre myst√®re, alors le joueur gagne la partie.

Dans le cas d'une victoire, vous devez indiquer au joueur combien d'essais lui ont √©t√© n√©cessaire pour gagner.

Si le joueur ne trouve pas le nombre myst√®re avec les 5 essais disponibles, il perd la partie.

```python
import random

#Le nombre al√©atoire compris entre 0 et 100 g√©n√©r√© par l'ordinateur
nombre_aleatoire = random.randint(0, 100)
#Nombre d'√©ssais total
nombre_essais = 5

#Nombre d'essais effectu√©s
nombre_essais_effectues = 0


#D√©but du Jeu myst√®re
print(f"***Le jeu du nombre myst√®re***")


for i in range(nombre_essais):
    #Nombre d'essais restants
    nombre_essais_restants = nombre_essais - nombre_essais_effectues
    
    print(f"Il vous reste {nombre_essais_restants} essais")
    nombre_saisi = input("Deviner le nombre : ")
    
    nombre_essais_effectues += 1

    
    while not (nombre_saisi.isdigit() and 0 <= int(nombre_saisi) <=100):
        print("Veuillez entrer un nombre valide (un entier entre 0 et 100)")
        print(f"Il vous reste {nombre_essais_restants} essais")
        nombre_saisi = input("Deviner le nombre : ")
    
    #Conversion en int du nombre_saisi
    nombre_saisi = int(nombre_saisi)
    
    if nombre_saisi == nombre_aleatoire:
        print (f"Bravo ! le nombre myst√®re √©tait {nombre_aleatoire}")
        print(f"Vous avez trouv√© ce nombre en {nombre_essais_effectues} essais")
        print("Fin du jeu !")
        break
        
    elif nombre_saisi > nombre_aleatoire:
        print(f"Le nombre myst√®re est plus petit que {nombre_saisi}")
        
    else:
        print(f"Le nombre myst√®re est plus grand que {nombre_saisi}")
            
if nombre_essais_effectues == nombre_essais and nombre_saisi != nombre_aleatoire:
    print(f"Dommage ! Le nombre myst√®re √©tait {nombre_aleatoire}")
    print("Fin du jeu")
```
